```python
import tensorflow as tf
import numpy as np
```

```python
inputs = np.ones([10, 4, 1]).astype(np.float32)
print(inputs)
inputs = tf.convert_to_tensor(inputs, dtype=tf.float32)

inputs
```

```python
states = np.ones([10, 4, 4]).astype(np.float32) * 1.5
print(states)
states = tf.convert_to_tensor(states, dtype=tf.float32)
states
```

```python
rnn_cell = tf.keras.layers.SimpleRNNCell(4)
```

```python
rnn_cell(inputs = inputs, states = states)
```
