# Example of data

To see, how they move through date

```python
def show_time_salecount_by_cat(sample_num = 5):
    fig = plt.figure(figsize=(20,9))
    ax1 = fig.add_subplot(3,1,1)
    ax2 = fig.add_subplot(3,1,2)
    ax3 = fig.add_subplot(3,1,3)
    hobby_sample = main[main.cat_id=="HOBBIES"].iloc[sample_num].values[6:]
    household_sample = main[main.cat_id=="HOUSEHOLD"].iloc[sample_num].values[6:]
    food_sample = main[main.cat_id=="FOODS"].iloc[sample_num].values[6:]
    ax1.plot(hobby_sample, color="r", label="Hobbies")
    ax2.plot(household_sample, color="b", label="HOUSEHOLD")
    ax3.plot(food_sample, color="g", label="FOODS")
    ax1.legend()
    ax2.legend()
    ax3.legend()
    plt.show()
```

```python
show_time_salecount_by_cat(sample_num = 5)
```

# EDA

# 1. Cat.1 / Sales count

```python
cat_salescount_table = main[["cat_id"]+list(main.columns[6:])].groupby(["cat_id"]).sum()
cat_salescount_table_val = cat_salescount_table.values
```

```python
fig = plt.figure(figsize=(20,9))
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)
ax1.plot(cat_salescount_table_val[0], color="r", label="FOODS")
ax2.plot(cat_salescount_table_val[1], color="b", label="HOBBIES")
ax3.plot(cat_salescount_table_val[2], color="g", label="HOUSEHOLD")
ax1.legend()
ax2.legend()
ax3.legend()
plt.show()
```

# 2. states / salescount

```python
states_salescount_table = main[["state_id"]+list(main.columns[6:])].groupby(["state_id"]).sum()
states_salescount_table_val = states_salescount_table.values
fig = plt.figure(figsize=(20,9))
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)
ax1.plot(states_salescount_table_val[0], color="r", label="CA")
ax2.plot(states_salescount_table_val[1], color="b", label="TX")
ax3.plot(states_salescount_table_val[2], color="g", label="WI")
ax1.legend()
ax2.legend()
ax3.legend()
plt.show()
```

# 3. Cat.1 / sales_counts + (unit time=Weekday)

```python
cat_list = ["FOODS","HOBBIES","HOUSEHOLD"]
tmp_wday_d = cal[["wday","d"]]

ii=0
for cat in cat_list:
    tmp_df = cat_salescount_table.T[[cat]]
    merge_df = pd.merge(tmp_df, tmp_wday_d, left_index=True, right_on="d", how="left")
    tmp_group = merge_df[["wday",cat]].groupby("wday").sum()
    print("Category is == {}".format(cat))
    fig = plt.figure(figsize=(20,3))
    ax1 = fig.add_subplot(1,1,1)
    ax1.plot(tmp_group.values.reshape(-1), color="b", label=cat)
    ax1.legend()
    plt.show()
    ii+=1
```

# 4. States / sales_counts + (unit time=Weekday)

```python
cat_list = ["CA","TX","WI"]
tmp_wday_d = cal[["wday","d"]]

ii=0
for cat in cat_list:
    tmp_df = states_salescount_table.T[[cat]]
    merge_df = pd.merge(tmp_df, tmp_wday_d, left_index=True, right_on="d", how="left")
    tmp_group = merge_df[["wday",cat]].groupby("wday").sum()
    print("Category is == {}".format(cat))
    fig = plt.figure(figsize=(20,3))
    ax1 = fig.add_subplot(1,1,1)
    ax1.plot(tmp_group.values.reshape(-1), color="b", label=cat)
    ax1.legend()
    plt.show()
    ii+=1
```

# 5. Cat.2 / sales count
