# Preprocessing

1. 지하철 및 버스에 Null 포함 값 삭제


```python
print("Before : ", train_df.shape)
train_df = train_df[np.logical_and(train_df["도보 10분거리 내 지하철역 수(환승노선 수 반영)"].notnull(),train_df["도보 10분거리 내 버스정류장 수"].notnull())]
print("After : ", train_df.shape)

targets_arr = []
inputs_arr = []
ap_idx_arr = []

area_list = list(train_df["지역"].unique())
supp_list = list(train_df["공급유형"].unique())
qual_list = list(train_df["자격유형"].unique())

for ap_code in train_df["단지코드"].unique():

    inputs_ = np.zeros(6+len(supp_list)+len(qual_list)+3)
    
    ap_idx_arr.append(ap_code)
    sample_df = train_df[train_df["단지코드"]==ap_code].reset_index(drop=True)
    sample_df.fillna(0, inplace=True)
    
    # Target Value
    target = sample_df["등록차량수"].values[0] / (sample_df["총세대수"].values[0] - sample_df["공가수"].values[0])
    targets_arr.append(target)
    
    # Input part
    tot_house_num = sample_df["전용면적별세대수"].sum()
    tot_house_num_add_install = sample_df["총세대수"].values[0]
    space = 0
    lent_amount = 0
    lent_fee = 0
    
    inputs_[0] = sample_df[sample_df["임대건물구분"]=="아파트"]["전용면적별세대수"].sum() / tot_house_num_add_install
    inputs_[1] = sample_df[sample_df["임대건물구분"]=="상가"]["전용면적별세대수"].sum() / tot_house_num_add_install
    inputs_[2] = (tot_house_num_add_install-tot_house_num) / tot_house_num_add_install
    inputs_[3] = area_list.index(sample_df["지역"].values[0])
    inputs_[4] = sample_df["도보 10분거리 내 지하철역 수(환승노선 수 반영)"].values[0]
    inputs_[5] = sample_df["도보 10분거리 내 버스정류장 수"].values[0]
    
    for row_idx in range(sample_df.shape[0]):
        row_samp = sample_df.iloc[row_idx].copy()
        inputs_[supp_list.index(row_samp["공급유형"])+6] +=  row_samp["전용면적별세대수"]/tot_house_num
        inputs_[qual_list.index(row_samp["자격유형"])+len(supp_list)+6] +=  row_samp["전용면적별세대수"]/tot_house_num
        space += row_samp["전용면적"] * row_samp["전용면적별세대수"]
        
        if row_samp["임대보증금"] == "-":
            lent_amount += 0
        else:
            lent_amount += (float(row_samp["임대보증금"])/1000) * row_samp["전용면적별세대수"]
        if row_samp["임대료"] == "-":
            lent_fee += 0
        else:
            lent_fee += (float(row_samp["임대료"])/1000) * row_samp["전용면적별세대수"]
    
    
    space = space/tot_house_num
    lent_amount = (lent_amount/tot_house_num)*1000
    lent_fee = (lent_fee/tot_house_num)*1000
    inputs_[-3] = space
    inputs_[-2] = lent_amount
    inputs_[-1] = lent_fee
    
    inputs_arr.append(inputs_)
        
        
    
targets_arr = np.array(targets_arr)
inputs_arr = np.array(inputs_arr)
ap_idx_arr = np.array(ap_idx_arr)

print(targets_arr.shape, inputs_arr.shape, ap_idx_arr.shape)
```

# LGBM Fitting

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
```

```python
cols = ["apart_rate", "market_rate", "install_rate", "area", "metro", "bus"] + supp_list + qual_list + ["space", "lent_amount", "lent_fee"]

inputs_df = pd.DataFrame(inputs_arr, columns = cols)
print(inputs_df.info())
x_train, x_valid, y_train, y_valid = train_test_split(inputs_df, targets_arr, test_size=0.30, random_state=7)

print(x_train.shape, y_train.shape)
print(x_valid.shape, y_valid.shape)

max_for_reg = np.ones(33)
min_for_reg = np.zeros(33)

max_for_reg[4:6] = x_train.max(axis=0)[4:6]
max_for_reg[-3:] = x_train.max(axis=0)[-3:]

min_for_reg[4:6] = x_train.min(axis=0)[4:6]
min_for_reg[-3:] = x_train.min(axis=0)[-3:]

x_train_norm = (x_train-min_for_reg) / (max_for_reg-min_for_reg)
x_valid_norm = (x_valid-min_for_reg) / (max_for_reg-min_for_reg)
```

```python
d_train = lgb.Dataset(x_train_norm, label=y_train)
d_valid = lgb.Dataset(x_valid_norm, label=y_valid)
watchlist = [d_valid]

params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': 'l2',
    'max_depth': -1,
    'early_stopping_rounds' : 20,
    'learning_rate': 0.01,
    'verbose': 0
    }
n_estimators = 1000


model = lgb.train(params, d_train, n_estimators, watchlist, categorical_feature=["area"], verbose_eval=1)

preds = model.predict(x_valid)
```

# Modeling

```python
print("Before : ", train_df.shape)
train_df = train_df[np.logical_and(train_df["도보 10분거리 내 지하철역 수(환승노선 수 반영)"].notnull(),train_df["도보 10분거리 내 버스정류장 수"].notnull())]
print("After : ", train_df.shape)


area_list = list(train_df["지역"].unique())
supp_list = list(train_df["공급유형"].unique())
qual_list = list(train_df["자격유형"].unique())


def return_preprocessed_dataset(df, train_set=True, col_select=[]):
    # 초기화 변수
    inputs_arr = []
    ap_idx_arr = []
    targets_arr = []
    for_multiple = []
    
    for ap_code in df["단지코드"].unique():
        # 빈 array 만들기
        inputs_ = np.zeros(6+len(supp_list)+len(qual_list)+3)
        
        # 인덱스 및 데이터프레임 추려내기
        ap_idx_arr.append(ap_code)
        sample_df = df[df["단지코드"]==ap_code].reset_index(drop=True)
        sample_df.fillna(0, inplace=True)
        
        # For Result
        for_multiple.append(sample_df["총세대수"].values[0] - sample_df["공가수"].values[0])
        
        if train_set:
            target = sample_df["등록차량수"].values[0] / (sample_df["총세대수"].values[0] - sample_df["공가수"].values[0])
            targets_arr.append(target)
            
        ### --- Input용 데이터 처리 시작
        tot_house_num = sample_df["전용면적별세대수"].sum()
        tot_house_num_add_install = sample_df["총세대수"].values[0]
        space = 0
        lent_amount = 0
        lent_fee = 0
        
        # 데이터 대입
        inputs_[0] = sample_df[sample_df["임대건물구분"]=="아파트"]["전용면적별세대수"].sum() / tot_house_num_add_install
        inputs_[1] = sample_df[sample_df["임대건물구분"]=="상가"]["전용면적별세대수"].sum() / tot_house_num_add_install
        inputs_[2] = (tot_house_num_add_install-tot_house_num) / tot_house_num_add_install
        inputs_[3] = area_list.index(sample_df["지역"].values[0])
        inputs_[4] = sample_df["도보 10분거리 내 지하철역 수(환승노선 수 반영)"].values[0]
        inputs_[5] = sample_df["도보 10분거리 내 버스정류장 수"].values[0]
        
        for row_idx in range(sample_df.shape[0]):
            row_samp = sample_df.iloc[row_idx].copy()
            inputs_[supp_list.index(row_samp["공급유형"])+6] +=  row_samp["전용면적별세대수"]/tot_house_num
            if row_samp["자격유형"] == 0:
                inputs_[qual_list.index("O")+len(supp_list)+6] +=  row_samp["전용면적별세대수"]/tot_house_num
            else:
                inputs_[qual_list.index(row_samp["자격유형"])+len(supp_list)+6] +=  row_samp["전용면적별세대수"]/tot_house_num
            space += row_samp["전용면적"] * row_samp["전용면적별세대수"]

            if row_samp["임대보증금"] == "-":
                lent_amount += 0
            else:
                lent_amount += (float(row_samp["임대보증금"])/1000) * row_samp["전용면적별세대수"]
            if row_samp["임대료"] == "-":
                lent_fee += 0
            else:
                lent_fee += (float(row_samp["임대료"])/1000) * row_samp["전용면적별세대수"]
                
        space = space/tot_house_num
        lent_amount = (lent_amount/tot_house_num)*1000
        lent_fee = (lent_fee/tot_house_num)*1000
        inputs_[-3] = space
        inputs_[-2] = lent_amount
        inputs_[-1] = lent_fee

        inputs_arr.append(inputs_)
    
    targets_arr = np.array(targets_arr)
    inputs_arr = np.array(inputs_arr)
    ap_idx_arr = np.array(ap_idx_arr)
    for_multiple = np.array(for_multiple)
    
    print(targets_arr.shape, inputs_arr.shape, ap_idx_arr.shape, for_multiple.shape)
    
    cols = ["apart_rate", "market_rate", "install_rate", "area", "metro", "bus"] + supp_list + qual_list + ["space", "lent_amount", "lent_fee"]

    inputs_df = pd.DataFrame(inputs_arr, columns = cols)
    
    if len(col_select)>0:
        inputs_df = inputs_df[col_select]
    
    print(inputs_df.info())
    
    return inputs_df, targets_arr, ap_idx_arr, for_multiple
```

```python
print("Start Train df preprocessing")
train_inputs, train_targets, train_ap_idx, train_for_multiple = return_preprocessed_dataset(df=train_df, train_set=True, col_select=["space", "lent_amount", "A", "area", "lent_fee", "영구임대", "국민임대", "bus", "metro"])

print()
print("Start Test df preprocessing")
test_df = pd.read_csv("./data/test.csv")
print(test_df.shape)
test_inputs, test_targets, test_ap_idx, test_for_multiple = return_preprocessed_dataset(df=test_df, train_set=False, col_select=["space", "lent_amount", "A", "area", "lent_fee", "영구임대", "국민임대", "bus", "metro"])
```

```python
train_data_idx = np.random.permutation(len(train_targets))
test_set_rate = 0.2
train_set_rate = 1-test_set_rate

for_train_idx = train_data_idx[:int(len(train_data_idx)*train_set_rate)]
for_valid_idx = train_data_idx[int(len(train_data_idx)*train_set_rate):]

x_train = train_inputs.iloc[for_train_idx]
y_train = train_targets[for_train_idx]
train_part_ap_idx = train_ap_idx[for_train_idx]
train_part_multiple = train_for_multiple[for_train_idx]

x_valid = train_inputs.iloc[for_valid_idx]
y_valid = train_targets[for_valid_idx]
valid_part_ap_idx = train_ap_idx[for_valid_idx]
valid_part_multiple = train_for_multiple[for_valid_idx]

print(x_train.shape, y_train.shape)
print(x_valid.shape, y_valid.shape)
```

```python
def make_arr_for_norm(train_df, col_need):
    max_for_reg = np.ones(train_df.shape[1])
    min_for_reg = np.zeros(train_df.shape[1])
    
    for idx, val, in enumerate(train_df.columns):
        if val in col_need:
            max_for_reg[idx] = train_df[val].values.max()
            min_for_reg[idx] = train_df[val].values.min()
    return max_for_reg, min_for_reg

max_for_reg, min_for_reg = make_arr_for_norm(train_df=x_train, col_need=["space", "lent_amount", "lent_fee", "bus", "metro"])

x_train_norm = (x_train-min_for_reg) / (max_for_reg-min_for_reg)
x_valid_norm = (x_valid-min_for_reg) / (max_for_reg-min_for_reg)
```

```python
d_train = lgb.Dataset(x_train_norm, label=y_train)
d_valid = lgb.Dataset(x_valid_norm, label=y_valid)
watchlist = [d_valid]

params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': 'l2',
    'max_depth': -1,
    'early_stopping_rounds' : 20,
    'learning_rate': 0.01,
    'verbose': 0
    }
n_estimators = 1000


model_use = lgb.train(params, d_train, n_estimators, watchlist, categorical_feature=["area"], verbose_eval=1)

train_preds = model_use.predict(x_train_norm)
valid_preds = model_use.predict(x_valid_norm)
```

```python
train_preds_transformed = train_preds*train_part_multiple
valid_preds_transformed = valid_preds*valid_part_multiple
```

```python
train_answer = []
for ap_code in train_part_ap_idx:
    ans = train_df[train_df["단지코드"]==ap_code]["등록차량수"].values[0]
    train_answer.append(ans)
train_answer = np.array(train_answer)

valid_answer = []
for ap_code in valid_part_ap_idx:
    ans = train_df[train_df["단지코드"]==ap_code]["등록차량수"].values[0]
    valid_answer.append(ans)
valid_answer = np.array(valid_answer)
```

```python
train_mae = sum(abs(train_preds_transformed - train_answer)) / len(train_preds_transformed)
valid_mae = sum(abs(valid_preds_transformed - valid_answer)) / len(valid_preds_transformed)

print("MAE of Train : {} / Valid : {}".format(train_mae, valid_mae))
```

```python
def ploting_standcol_with_cars_num(df):
    for col in df.columns:
        if col == "등록차량수":
            break
        elif col == "단지코드":
            continue
        elif col in ["총세대수", "지역", "공가수", "도보 10분거리 내 지하철역 수(환승노선 수 반영)", "도보 10분거리 내 버스정류장 수", "단지내주차면수"]:
            print(col)
            vals_arr = []
            cars_nums_arr = []
            for ap_code in df["단지코드"].unique():
                sample_df = df[df["단지코드"]==ap_code]

                val_1 = sample_df[col].values[0]
                cars_num = sample_df["등록차량수"].values[0]
                vals_arr.append(val_1)
                cars_nums_arr.append(cars_num)
            
            vals_arr = np.array(vals_arr)
            cars_nums_arr = np.array(cars_nums_arr)
            fig = plt.figure(figsize=(5,5))
            ax = fig.add_subplot(111)
            ax.scatter(vals_arr, cars_nums_arr)
            plt.show()
            
            
def low_class(idx_arr, targets, for_find_idx_arr, col_list):
    val_list = []
    target_arr = []
    for idx, val in enumerate(idx_arr):
        if val in for_find_idx_arr:
            ser = inputs_tot.iloc[idx].values
            val_list.append(ser)
            target_arr.append(targets[idx])
    val_arr = np.array(val_list)
    target_arr = np.array(target_arr)
    return pd.DataFrame(val_arr, columns=col_list), target_arr

def ploting_for_low_class(bellow_df, above_df):
    for col in bellow_df.columns:
        print("-----Columns name ------ :",col)
        bellow_vals_arr = bellow_df[col].values
        above_val_arr = above_df[col].values
        
        fig = plt.figure(figsize=(10,5))
        ax_1 = fig.add_subplot(121)
        ax_2 = fig.add_subplot(122)
        ax_1.boxplot(bellow_vals_arr)
        ax_2.boxplot(above_val_arr)
        ax_1.set_title("Bellow")
        ax_2.set_title("Above")
        plt.show()
```
