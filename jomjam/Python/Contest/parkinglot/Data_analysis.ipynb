```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import math
%matplotlib inline
import matplotlib
matplotlib.rc('axes', unicode_minus=False)
```

# 하위 층으로 분석

```python
train_df = pd.read_csv("./data/train.csv")
print(train_df.shape)
```

### 상위 층 가중치로 만들기

#### 1. 지역

```python
areas_arr = []
car_num_by_household_arr = []
for ap_code in train_df["단지코드"].unique():
    sample_df = train_df[train_df["단지코드"]==ap_code]
    tot_house_num = sample_df["총세대수"].values[0] - sample_df["공가수"].values[0]
    cars_num = sample_df["등록차량수"].values[0]
    car_num_by_household = cars_num/tot_house_num
    
    areas_arr.append(sample_df["지역"].values[0])
    car_num_by_household_arr.append(car_num_by_household)
    
areas_arr = np.array(areas_arr)
car_num_by_household_arr = np.array(car_num_by_household_arr)

cars_num_by_household_by_area = {}

# 지역별 평균 값을 보자
for area_ in np.unique(areas_arr):
    mean_val = car_num_by_household_arr[areas_arr==area_].mean()
    cars_num_by_household_by_area[area_] = mean_val
    
print(cars_num_by_household_by_area)
```

#### 2. 지하철

```python
metro_num_arr = []
################################################# 세대별 구하는 파트 #################################################
car_num_by_household_arr = []
for ap_code in train_df["단지코드"].unique():
    sample_df = train_df[train_df["단지코드"]==ap_code]
    tot_house_num = sample_df["총세대수"].values[0] - sample_df["공가수"].values[0]
    cars_num = sample_df["등록차량수"].values[0]
    car_num_by_household = cars_num/tot_house_num
    car_num_by_household_arr.append(car_num_by_household)
##################################################################################################################
    
    if math.isnan(sample_df["도보 10분거리 내 지하철역 수(환승노선 수 반영)"].values[0]):
        metro_num_arr.append(-1)
    else:
        metro_num_arr.append(sample_df["도보 10분거리 내 지하철역 수(환승노선 수 반영)"].values[0])
    
metro_num_arr = np.array(metro_num_arr)
car_num_by_household_arr = np.array(car_num_by_household_arr)

cars_num_by_household_by_metro = {}

# 지하철 개수 별 평균 값을 보자
for metro_num_ in np.unique(metro_num_arr):
    mean_val = car_num_by_household_arr[metro_num_arr==metro_num_].mean()
    cars_num_by_household_by_metro[metro_num_] = mean_val
    
print(cars_num_by_household_by_metro)
```

#### 3. 버스

```python
bus_num_arr = []
################################################# 세대별 구하는 파트 #################################################
car_num_by_household_arr = []
for ap_code in train_df["단지코드"].unique():
    sample_df = train_df[train_df["단지코드"]==ap_code]
    tot_house_num = sample_df["총세대수"].values[0] - sample_df["공가수"].values[0]
    cars_num = sample_df["등록차량수"].values[0]
    car_num_by_household = cars_num/tot_house_num
    car_num_by_household_arr.append(car_num_by_household)
##################################################################################################################
    
    if math.isnan(sample_df["도보 10분거리 내 버스정류장 수"].values[0]):
        bus_num_arr.append(-1)
    else:
        bus_num_arr.append(sample_df["도보 10분거리 내 버스정류장 수"].values[0])
    
bus_num_arr = np.array(bus_num_arr)
car_num_by_household_arr = np.array(car_num_by_household_arr)

cars_num_by_household_by_bus = {}

# 버스 개수 별 평균 값을 보자
for bus_num_ in np.unique(bus_num_arr):
    mean_val = car_num_by_household_arr[bus_num_arr==bus_num_].mean()
    cars_num_by_household_by_bus[bus_num_] = mean_val
    
print(cars_num_by_household_by_bus)
```

#### 4. 단지내주차면수, 총세대수
이 두가지는 세대당 주차 가능 대수로 치환해서 사용

#### 5. 공가구수
사실 이 수치는 집이 비어있는게 상관이 있을 듯 싶음...   
같은 단지내에 집이 비어 있다고 차를 더 사거나 더 등록하는 경우는 내가 생각하기에는 없음  
이 값은 실직 차 등록 대수와 더 직접적 관련이 있는 듯함  
이 값을 활용해서 실질 세대당 등록차량 수를 구해서 이를 Target 값으로 활용해보자

### 상위층 값 제거

```python
train_copy = train_df.copy()
train_copy["실질거주세대수"] = train_copy["총세대수"] - train_copy["공가수"]
train_copy["세대당등록차량수"] = train_copy["등록차량수"] / train_copy["실질거주세대수"]

def area_transform(raw):
    return cars_num_by_household_by_area[raw["지역"]]
def metro_transform(raw):
    if math.isnan(raw["도보 10분거리 내 지하철역 수(환승노선 수 반영)"]):
        return cars_num_by_household_by_metro[-1]
    else:
        return cars_num_by_household_by_metro[raw["도보 10분거리 내 지하철역 수(환승노선 수 반영)"]]
def bus_transform(raw):
    if math.isnan(raw["도보 10분거리 내 버스정류장 수"]):
        return cars_num_by_household_by_bus[-1]
    else:
        return cars_num_by_household_by_bus[raw["도보 10분거리 내 버스정류장 수"]]

train_copy["지역별_세대당등록차량수"] = train_copy.apply(area_transform, axis=1)
train_copy["지하철별_세대당등록차량수"] = train_copy.apply(metro_transform, axis=1)
train_copy["버스별_세대당등록차량수"] = train_copy.apply(bus_transform, axis=1)
```

```python
train_copy["세대당_주차가능"] = train_copy["단지내주차면수"] / train_copy["총세대수"]
train_copy['law_layer_target'] = ((train_copy["세대당등록차량수"] / train_copy["지역별_세대당등록차량수"]) / train_copy["지하철별_세대당등록차량수"]) / train_copy["버스별_세대당등록차량수"]
train_copy.head()
```

일단은 제외하는 방향으로 생각

```python
need_df_2 = need_df.dropna()
need_df_2 = need_df_2[np.logical_and(need_df_2["임대료"]!="-", need_df_2["임대보증금"]!="-")].reset_index()

target_arr = need_df_2["law_layer_target"].values
continuos_df = need_df_2[["전용면적","임대료","임대보증금"]].astype("float")
continuos_df_min = continuos_df.min(axis=0)
continuos_df_max = continuos_df.max(axis=0)
continuos_df_range = continuos_df_max - continuos_df_min
continuos_df = (continuos_df - continuos_df_min) / continuos_df_range

result_df = pd.concat([pd.get_dummies(need_df_2[["임대건물구분","자격유형","공급유형"]]), continuos_df], axis=1)
result_df = pd.concat([result_df,need_df_2[["세대당_주차가능"]]], axis=1)

print(result_df.shape)
result_df
```
