# Regression

### Preprocessing

```python
from collections import deque
```

```python
sample_0_train_part = train_part[train_part.sample_id==0][["open"]]
sample_0_target_part = target_part[target_part.sample_id==0][["open"]]

print(sample_0_train_part.shape, sample_0_target_part.shape)
```

```python
sample_0_train_part.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

plt.figure(figsize=(10,3))
plt.plot(np.arange(1380), sample_0_train_part.open.values)
plt.show()
```

```python
x=[]
y=[]
for idx in range(sample_0_train_part.shape[0]):
    if idx < 6:
        continue
    part_data = sample_0_train_part.iloc[idx-6:idx].diff(periods=1, axis=0).iloc[1:].values.reshape(-1)
    target_data = sample_0_train_part.iloc[idx].values[0] - sample_0_train_part.iloc[idx-1].values[0]
    
    x.append(part_data)
    y.append(target_data)
    
x = np.array(x)
y = np.array(y)
print(x.shape, y.shape)
```

### Modeling

```python
from sklearn.linear_model import LinearRegression
```

```python
line_fitter = LinearRegression()
line_fitter.fit(x,y)
```

```python
train_predict = line_fitter.predict(x)

plt.figure(figsize=(15,5))
plt.plot(np.arange(1374), y)
plt.plot(np.arange(1374), train_predict)
plt.show()
```

```python
def diff_arr(arr):
    new_arr = []
    for idx in range(len(arr)-1):
        new_arr.append(arr[idx+1]-arr[idx])
    return np.array(new_arr)

preds = []
for ii in range(0,120):
    if ii == 0:
        inputs = sample_0_train_part.iloc[-6:].diff(periods=1, axis=0).iloc[1:].values.reshape(-1).reshape(1,5)
        pred = line_fitter.predict(inputs)
        preds.append(pred[0])
    
    elif ii<6:
        inputs = sample_0_train_part.iloc[-6+ii:].values.reshape(-1)
        inputs = diff_arr(inputs)
        new_inputs = np.concatenate((inputs, np.array(preds)), axis=0)
        pred = line_fitter.predict(new_inputs.reshape(1,5))
        preds.append(pred[0])
        
    else:
        inputs = np.array(preds[-5:])
        pred = line_fitter.predict(inputs.reshape(1,5))
        preds.append(pred[0])
```

```python
print(len(preds))
```

```python
last_open_val = sample_0_train_part.iloc[-1].values[0]

trans_data = []
for i in range(120):
    if i == 0:
        trans_data.append(last_open_val+preds[0])
    else:
        last_data = trans_data[-1]
        trans_data.append(last_data + preds[i])
print(len(trans_data))
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.arange(1380), sample_0_train_part.values, color="b")
plt.plot(np.arange(1380, 1500), sample_0_target_part.values.reshape(-1), color="g")
plt.plot(np.arange(1380, 1500), trans_data, color="r")
plt.show()
```

### Evaluation

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_breusch_godfrey
import scipy.stats as stats

target_diff = sample_0_target_part.diff(periods=1).iloc[1:].values.reshape(-1)
preds_start_from_1 = preds[1:]

residual = target_diff - preds_start_from_1

fig = plt.figure(figsize=(20,10))
ax_1 = fig.add_subplot(3,1,1)
ax_2 = fig.add_subplot(3,1,2)
ax_3 = fig.add_subplot(3,1,3)

ax_1.plot(np.arange(len(residual)), residual)
plot_acf(residual, ax=ax_2)
ax_3.hist(residual, bins=50)

plt.show()

def return_ac(arr, lag):
    mean_val = arr.mean()
    r_val = ((arr-mean_val)[:-lag]*(np.roll(arr,-lag)-mean_val)[:-lag]).sum() / np.power(arr-mean_val, 2).sum()
    return r_val

def ljung_box(arr, k):
    t = len(arr)
    h = int(t/5)
    sum_num = 0
    for i in range(1, h+1):
        r_i = return_ac(arr=arr, lag=i)
        sum_num += ((r_i*r_i)/(t-i))
    q = t*(t+2)*sum_num
    degree_of_freedom = h-k
    chi2 = np.linspace(0.5, 50, 100)
    pdf_arr = stats.chi2(degree_of_freedom).pdf(chi2)
    return pdf_arr[chi2>=q].sum()

print(ljung_box(np.array(preds), k=5))
```
