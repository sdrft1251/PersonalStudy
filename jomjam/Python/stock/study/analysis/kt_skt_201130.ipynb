# Data Load

```python
import pandas as pd
```

```python
skt = pd.read_csv("/home/jaemincho/works/stock/study/data/A017670.csv")
kt = pd.read_csv("/home/jaemincho/works/stock/study/data/A030200.csv")

skt["datetime"] = skt['date'].astype(str)
skt["datetime"] = pd.to_datetime(skt['datetime'])
skt["datetime"] = skt["datetime"].dt.tz_localize('UTC').dt.strftime("%Y/%m/%d")
skt["datetime"] = pd.to_datetime(skt['datetime'])

kt["datetime"] = kt['date'].astype(str)
kt["datetime"] = pd.to_datetime(kt['datetime'])
kt["datetime"] = kt["datetime"].dt.tz_localize('UTC').dt.strftime("%Y/%m/%d")
kt["datetime"] = pd.to_datetime(kt['datetime'])

skt.index = skt["datetime"]
kt.index = kt["datetime"]

skt_after_2010 = skt[skt.date>=20100101]
kt_after_2010 = kt[kt.date>=20100101]

print(skt_after_2010.shape, kt_after_2010.shape)
```

# Total set

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib.dates as mdates

months = mdates.MonthLocator()
year_months_format = mdates.DateFormatter('%Y/%m')


fig = plt.figure(figsize=(40,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(skt_after_2010.datetime, skt_after_2010.high, color="g")
ax1.plot(skt_after_2010.low, color="y")
ax1.plot(skt_after_2010.close, color="r")
ax2.plot(kt_after_2010.high, color="g")
ax2.plot(kt_after_2010.low, color="y")
ax2.plot(kt_after_2010.close, color="r")

ax1.xaxis.set_major_locator(months)
ax1.xaxis.set_major_formatter(year_months_format)
ax2.xaxis.set_major_locator(months)
ax2.xaxis.set_major_formatter(year_months_format)

ax1.grid(True)
ax2.grid(True)

ax1.tick_params(labelrotation=45)
ax2.tick_params(labelrotation=45)

plt.show()
```

### Note

- 4G Start date = 20110701(SKT, LGU+) 201201(KT)
