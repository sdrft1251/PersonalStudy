```python
from src import portfolio
from src.utils import get_stock_code_list
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rc('axes', unicode_minus=False)
from src.utils import return_data
from src.portfolio import select_code, return_portfolio
from src.backtesting import test_start_func
```

```python
start_time = datetime(year=2018, month=1, day=1)
result_rate = 1

while start_time+timedelta(days=180) < datetime(year=2021, month=6, day=20):
    get_data_start_datetime = start_time
    get_data_end_datetime = start_time + timedelta(days=90)
    test_end_datetime = get_data_end_datetime + timedelta(days=15)
    print(get_data_start_datetime.strftime("%Y-%m-%d"), get_data_end_datetime.strftime("%Y-%m-%d"), test_end_datetime.strftime("%Y-%m-%d"))
    
    code_list, retain_list = return_portfolio(get_data_start_datetime=get_data_start_datetime, get_data_end_datetime=get_data_end_datetime,\
                                              short_rate=0.25, extract_code_num=20)
    profit_rate = test_start_func(code_list=code_list, retain_list=retain_list, test_start_datetime=get_data_end_datetime, test_end_datetime=test_end_datetime)
    # 수수료 제외
    result_rate = result_rate * profit_rate * 0.995
    print(test_end_datetime.strftime("%Y-%m-%d") + " Profit is : {}".format(result_rate))
    start_time += timedelta(days=15)
```
