2. 한 방향의 종목을 담았으면 그와 완전히 다른 방향으로 움직이는 종목을 하나 추가해야함.

현재는 서로 움직이는 방향을 알았으니, 위 목적성에 부합하는 종목을 찾아보자.

# 우 상향?

종목 별로 선형 회귀에 적합 후, 그 기울기를 데이터로 가져오자.

```python
from sklearn.linear_model import LinearRegression
```

```python
coef_dict = {}

for code in dropped_df.columns:
    price_by_code = dropped_df[code].values
    
    normed_price_by_code = price_by_code / price_by_code[0]
    
    line_fitter = LinearRegression()
    line_fitter.fit(np.arange(len(normed_price_by_code)).reshape(-1,1), normed_price_by_code)
    
    coef_val = line_fitter.coef_[0]
    coef_dict[code] = coef_val
```

```python
sorted_coef_dict = sorted(coef_dict.items(), key=(lambda x:x[1]), reverse=True)

sorted_coef_dict
```

상위 종목에 테마성이 보이는 종목이 너무 많음...  
어떻게 거르나...

> 테마성의 정의를 근거로 판단해보자.  
1. 가격이 급격하게 오름 -> 증가율이 너무 가파른 것 제외 (분위수 활용하면 되지 않을까)
2. 기존 가치보다 급격하게 증가 -> PER값과 같은 기준으로 보아야 하는데, 과거 시점의 테스트가 가능하려나...
3. 안정적인 목표 수익률 기준으로 판단해볼까? -> 목표 수익률 = 연 30% / 3개월이면, 약 7.5%를 기준으로..

위에서 1번과 3번을 테스트 해보자

```python
# 오른 것 중에 중간 부분
only_increase_rate = []
for key, val in sorted_coef_dict:
    only_increase_rate.append(val)
only_increase_rate = np.array(only_increase_rate)
only_increase_rate
```

```python
above_zero = only_increase_rate[only_increase_rate>0]
print(len(above_zero))

mid_idx = int(len(above_zero)/2)
# 중앙으로부터 좌우로 10개 정도?
extracted_code = sorted_coef_dict[mid_idx-20:mid_idx+20]
extracted_code
```

위 값들은 1일 평균 수익률 0.1% -> 연 240일로 보았을 때, 연 24퍼센트 수익률 나쁘지 않네...  
위 값들과 상관도는 완전 반대이나, 상승률이 나쁘지 않은 것으로 선정해야됨..

```python
results = []
for key, val in extracted_code:

    
    corr_ser = dropped_df.corr()[key].sort_values()
    cand_code = corr_ser[corr_ser<-0.5]
    
    otherside = {}
    for idx in cand_code.index:
        tmp = coef_dict[idx]
        if tmp>0.001 :
            otherside[idx] = tmp
    
    if len(otherside)>=1:
        sorted_otherside = sorted(otherside.items(), key=(lambda x:x[1]), reverse=True)
        
        results.append([key, val, sorted_otherside[0][0], sorted_otherside[0][1]])
        
result = np.array(results)
print(results)
```

위 과정을 묶어버리고, 옵션값을 넣게 끔 함수화 하자

```python
for_see_num = 60

above_zero = only_increase_rate[only_increase_rate>0]
mid_idx = int(len(above_zero)/2)
# 중앙으로부터 좌우로 10개 정도?
extracted_code = sorted_coef_dict[mid_idx-int(for_see_num*0.7):mid_idx+int(for_see_num*0.3)]
results = []
for key, val in extracted_code:
    corr_ser = dropped_df.corr()[key].sort_values()
    cand_code = corr_ser[corr_ser<-0.5]
    
    otherside = {}
    for idx in cand_code.index:
        tmp = coef_dict[idx]
        if tmp>0.001 :
            otherside[idx] = tmp
    
    if len(otherside)>=1:
        sorted_otherside = sorted(otherside.items(), key=(lambda x:x[1]), reverse=True)
        
        results.append([key, val, sorted_otherside[0][0], sorted_otherside[0][1]])
        
result = np.array(results)
print(len(results))
print(results)
```

# Back Testing 진행

### 포트폴리오 구성

```python
from src.backtesting import test_func
from src.utils import return_data
```

```python
can_buy = 1000000 * 0.1

code_list = []
retain_list = []
for arr in results:
    data_1 = return_data(arr[0], "2021-03-02", "2021-03-02").Close.values[0]
    data_2 = return_data(arr[2], "2021-03-02", "2021-03-02").Close.values[0]

    can_buy_num_1 = int(can_buy/data_1)
    can_buy_num_2 = int(can_buy/data_2)
    
    if can_buy_num_1>=1 and can_buy_num_2>=1:
        code_list.append(arr[0])
        code_list.append(arr[2])
        retain_list.append(can_buy_num_1*data_1)
        retain_list.append(can_buy_num_2*data_2)
    
    print(arr[0], data_1, can_buy_num_1, can_buy_num_1*data_1)
    print(arr[2], data_2, can_buy_num_2, can_buy_num_2*data_2)
```

```python
print(code_list)
print(retain_list)
```

```python
for_date_df = return_data("005930", "2021-03-02", "2021-06-20")
before_list = for_date_df.index[:-1]
after_list = for_date_df.index[1:]

port_ex = pd.DataFrame()
port_ex['code'] = code_list
port_ex[before_list[0]] = retain_list
print(port_ex)
```
