```python
from src.utils import return_data
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rc('axes', unicode_minus=False)
```

```python
ex_df = return_data("005930", "2010-01-01", "2021-06-20")
print(ex_df.shape)
ex_df.head()
```

# 추세? 계절성? 분석 해보자

```python
def for_delete_trend(arr, window):
    results = []
    tot_length = len(arr)
    for idx in range(tot_length-window):
        sample = arr[idx:idx+window]
        line_fitter = LinearRegression()
        line_fitter.fit(np.arange(window).reshape(-1,1), sample)
        pred_result = line_fitter.predict(np.arange(window)[-1].reshape(-1,1))
        results.append(pred_result[0])
    return np.array(results)
```

```python
window = 20
show_sample = 500


# 회귀 계수 출략 및 길이 조정
pred_results = for_delete_trend(ex_df.Close.values, window)
closed = ex_df.Close[window:]

delete_trend = closed - pred_results
```

```python
sample = delete_trend[:window]

fft = np.fft.fft(sample) / len(sample)
fft_magnitude = abs(fft)
```

```python
plt.subplot(2,1,1)
plt.plot(np.arange(window),sample)
plt.grid()
 
plt.subplot(2,1,2)
plt.stem(np.fft.fftshift(fft_magnitude))
plt.grid()                                                                                 
 
plt.show()
```
